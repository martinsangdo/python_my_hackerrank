```python
def get_high_value_orders(orders):
    high_value = []
    for order in orders:
        # BUG 1: Comparison with string instead of int/float
        if order['total'] > "100":
            high_value.append(order)
    
    # BUG 2: Trying to sort a list of dicts without a key/lambda
    # BUG 3: 'date' might be missing in some dicts (KeyError)
    high_value.sort() 
    
    return high_value

# Test Data
my_orders = [
    {'id': 1, 'total': 150, 'date': '2024-01-01'},
    {'id': 2, 'total': 50, 'date': '2024-01-02'},
    {'id': 3, 'total': 200}
]

print(get_high_value_orders(my_orders))
```
