```python
import heapq
```

```python
#read data from file
arr_data = []
with open('../data/big_array.txt', 'r') as f:
    for line in f:  #assume only 1 line
        arr_data = line.split(' ')
```

```python
digits = []
for c in arr_data:
    if (int(c) > 0):
        digits.append(int(c))
```

```python
#cookie questions in https://www.hackerrank.com/challenges/one-week-preparation-kit-jesse-and-cookies/problem
def find_cookie_mix_k(arr, k):
    heapq.heapify(arr)
    #print(arr)
    smallest = heapq.nsmallest(1, arr)
    num = 0
    while smallest[0] < k and len(arr) > 1:
        smallest, smallest2 = heapq.nsmallest(2, arr)
        #get 2 smallest items
        newSum = smallest + 2 * smallest2
        heapq.heappop(arr)
        heapq.heappop(arr)
        heapq.heappush(arr, newSum)
        smallest = heapq.nsmallest(1, arr)
        num += 1
        #print(arr)
    if heapq.nsmallest(1, arr)[0] >= k:
        return num
    return -1   #no solution
#test
k = 9
#arr = [1, 5, 2, 8, 3, 9, 4, 6, 2]
print(find_cookie_mix_k(digits, k)) #take 1 min

```

```python
# Push an element onto the heap
heapq.heappush(digits, 2)
print("Heap after push:", digits)

# Pop the smallest element from the heap
smallest = heapq.heappop(digits)
print("Smallest:", smallest)
print("Heap after pop:", digits)

#Find the 3 smallest elements.
print(heapq.nsmallest(3, digits))
print(heapq.nlargest(3, digits))
```

```python
#stack, use array
class MyStack:
    def __init__(self):
        self.arr = []
    
    def push(self, item):
        self.arr.append(item)
    
    def pop(self):
        self.arr.pop()

#test
stack = MyStack()
stack.push(9)
stack.push('5')
stack.push('t')
stack.pop()

print(stack.arr)
```
